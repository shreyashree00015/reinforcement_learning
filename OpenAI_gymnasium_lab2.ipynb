Q3. Running an environment-agent interaction cycle

```python
import gymnasium as gym
env = gym.make('CartPole-v1', render_mode='rgb_array')
import time
for _ in range(10):
  env.reset()
  done = False
  while not done:
    action = env.action_space.sample()
    obs,reward,done,x,info = env.step(action)
    env.render()
    time.sleep(0.01)
```

Q6. Alter the interaction cycle code to record all the rewards from each training episode and plot
it against the episode number using Matplotlib

```python
import pyvirtualdisplay
_display = pyvirtualdisplay.Display(visible=False,size=(1200, 900))
_ = _display.start()
!echo $DISPLAY
import gymnasium as gym
import matplotlib.pyplot as plt
from IPython import display
import time

env = gym.make('CartPole-v1', render_mode='rgb_array')
num_episodes = 100
cartpole_episode_rewards = []
for episode in range(num_episodes):
    env.reset()
    total_reward = 0
    done = False
    while not done:
        action = env.action_space.sample()
        obs,reward,done,x,info = env.step(action)
        total_reward += reward
    cartpole_episode_rewards.append(total_reward)

plt.plot(range(num_episodes), cartpole_episode_rewards, marker='o')
plt.xlabel('Episode Number')
plt.ylabel('Total Reward')
plt.title('Total Reward per Episode')
plt.grid(True)
plt.show()

env.close()
```

```python
import matplotlib.pyplot as plt

env = gym.make("MountainCar-v0")
mountaincar_episode_rewards = []
num_episodes = 100

for episode in range(num_episodes):
    obs = env.reset()
    total_reward = 0
    done = False
    while not done:
        action = env.action_space.sample()
        obs,reward,done,x,info = env.step(action)
        total_reward += reward

    mountaincar_episode_rewards.append(total_reward)

plt.plot(range(num_episodes), mountaincar_episode_rewards, marker='o')
plt.xlabel('Episode Number')
plt.ylabel('Total Reward')
plt.title('Total Reward per Episode')
plt.grid(True)
plt.show()

env.close()
```

```python
env = gym.make('LunarLander-v2')
lunarlander_episode_rewards = []
num_episodes = 100

for episode in range(num_episodes):
    env.reset()
    total_reward = 0
    done = False

    while not done:
        action = env.action_space.sample()
        obs,reward,done,x,info = env.step(action)
        total_reward += reward
    lunarlander_episode_rewards.append(total_reward)

plt.plot(range(num_episodes), lunarlander_episode_rewards, marker='o')
plt.xlabel('Episode Number')
plt.ylabel('Total Reward')
plt.title('Total Reward per Episode')
plt.grid(True)
plt.show()

env.close()
```

```python
plt.figure(figsize=(20, 5))

# Plotting CartPole rewards
plt.subplot(1, 3, 1)
plt.plot(range(num_episodes), cartpole_episode_rewards, marker='o')
plt.xlabel('Episode Number')
plt.ylabel('Total Reward')
plt.title('CartPole')

# Plotting MountainCar rewards
plt.subplot(1, 3, 2)
plt.plot(range(num_episodes), mountaincar_episode_rewards, marker='o')
plt.xlabel('Episode Number')
plt.ylabel('Total Reward')
plt.title('MountainCar')

# Plotting LunarLander rewards
plt.subplot(1, 3, 3)
plt.plot(range(num_episodes), lunarlander_episode_rewards, marker='o')
plt.xlabel('Episode Number')
plt.ylabel('Total Reward')
plt.title('LunarLander')

plt.tight_layout()
plt.show()
```
